# 📓 PANDUAN LENGKAP OPERASIONAL JUPYTER NOTEBOOK (.IPYNB)

Gunakan dokumen ini sebagai referensi utama saat belajar Python di lingkungan Notebook (VS Code/Jupyter).

---

## 🕹️ 1. DUA MODE UTAMA (WAJIB PAHAM)
Sebelum menekan shortcut, perhatikan warna garis/bingkai di sebelah kiri cell:

*   **MODE NAVIGASI (COMMAND MODE)**: Garis warna **Biru**.
    *   *Cara Masuk*: Tekan **`Esc`**.
    *   *Fungsi*: Mengelola kotak (Cell), menambah, menghapus, atau pindah-pindah.
*   **MODE NGETIK (EDIT MODE)**: Garis warna **Hijau** (atau ada kursor kedap-kedip).
    *   *Cara Masuk*: Tekan **`Enter`** pada cell yang dipilih.
    *   *Fungsi*: Menulis kode atau catatan di dalam kotak.

---

## ⌨️ 2. SHORTCUT SAKTI (COMMAND MODE - TEKAN `ESC`)

| Tombol | Fungsi | Penjelasan |
| :--- | :--- | :--- |
| **`A`** | **Above** | Tambah cell baru di **ATAS** |
| **`B`** | **Below** | Tambah cell baru di **BAWAH** |
| **`M`** | **Markdown** | Ubah kotak jadi **Catatan** (Teks) |
| **`Y`** | **Code** | Ubah kotak jadi **Koding** (Python) |
| **`D, D`** | **Delete** | Hapus cell (pencet D dua kali cepat) |
| **`Z`** | **Undo** | Batalkan penghapusan cell |
| **`L`** | **Line Numbers**| Munculkan/Sembunyikan nomor baris |
| **`O`** | **Output** | Sembunyikan/Munculkan hasil running |

---

## 🏃 3. CARA MENJALANKAN (RUNNING) CELL

| Kombinasi | Aksi | Kegunaan |
| :--- | :--- | :--- |
| **`Shift + Enter`** | Run + Pindah | Jalanin kode & lanjut ke bawah (Paling sering dipakai) |
| **`Ctrl + Enter`** | Run In Place | Jalanin kode & tetap di kotak yang sama |
| **`Alt + Enter`** | Run + New | Jalanin kode & otomatis bikin kotak baru di bawah |

---

## 🧠 4. LOGIKA "PINTAR" NOTEBOOK (STATE)

Notebook itu beda sama script biasa. Dia punya memori yang **berurutan**:
1.  **Variabel Global**: Kalau lo bikin `nama = "Ojay"` di cell atas dan lo RUN, cell paling bawah pun bakal kenal siapa itu `nama`.
2.  **Angka Index `[ ]`**: Perhatikan angka di sebelah kiri cell (misal `[1]`, `[2]`). Itu adalah urutan jalaninnya. **Bukan posisi atas-bawah yang menentukan, tapi siapa yang di-RUN duluan.**
3.  **Kernel Stuck `[*]`**: Kalau muncul tanda bintang, artinya Python lagi mikir keras.
    *   **Solusi**: Klik **Stop/Interrupt** atau **Restart Kernel** (Ikon panah muter).

---

## ✍️ 5. TIPS CATATAN MARKDOWN (VISUAL)
Ubah cell jadi Markdown (`M`), lalu gunakan format ini:

*   `# Judul Utama`
*   `## Sub Judul`
*   `**Teks Tebal**` atau `*Teks Miring*`
*   `- Poin List`
*   `1. Angka List`
*   Untuk blok kode: \`print("Halo")\`

---

## 🛠️ 6. FITUR VS CODE POWER-USER
*   **`Ctrl + /`** : Menjadikan baris kode sebagai komentar (Disable).
*   **`Tab`** : Auto-complete (Biar nggak typo ngetik variabel).
*   **`Shift + Tab`** : Intip penjelasan fungsi (Dokumentasi).
