# 🧪 Python Variables & Strings: The Deep Dive

## A. Variable Naming & Unpacking
Python sangat fleksibel dalam pengisian nilai. Gunakan `snake_case` untuk nama variabel.
- **Multiple Assignment**: `x, y, z = 1, 2, 3`
- **Unpacking**: Membongkar isi list ke variabel terpisah.

Variabel di Python itu unik karena lo nggak perlu deklarasi tipe data (kayak int atau str). Python langsung tahu sendiri pas lo isi nilainya.

```python
x = 5       # x jadi tipe angka (int)
y = "John"  # y jadi tipe teks (str)
```

2. Casting (Paksa Tipe Data)

Kalau lo mau mastiin sebuah variabel itu harus angka atau teks, lo bisa "paksa" pake fungsi:

x = str(3)    -> x jadi "3" (teks)

y = int(3)    -> y jadi 3 (angka bulat)

z = float(3)  -> z jadi 3.0 (angka desimal)

```python
print (float(x))
```

3. Get the Type (Cek Isi)

Pake fungsi type() buat liat Python nganggep variabel lo itu apa.

```python
x = 5
print(type(x)) # Hasilnya: <class 'int'>
```

Di Python, variabel itu bukan "kotak" yang nyimpen nilai, tapi "label" yang nempel ke objek di memori.
- Multi-Word Names: Karena kita nggak bisa pakai spasi, ada 3 gaya:
  - myVariableName (Camel Case)
  - MyVariableName (Pascal Case)
  - my_variable_name (Snake Case) -> Ini standar resmi Python (PEP 8).

Ini fitur "sakti" Python yang sering bikin bingung pemula:

- Many Values to Multiple Variables:
  - **Lo bisa isi banyak variabel sekaligus dalam satu baris** : 
  - ` x, y, z = "Orange", "Banana", "Cherry" `

```python
x,y,z = "aku", "kamu", "dia"
print(x)
print(y)
print(z)
```

- One Values to Multiple Variables:
  - **Satu nilai dibagi banyak variable** : 
  - ` x, y, z = "Orange"`

```python
x = y = z = "aku"
print(x)
print(y)
print(z)
```

```python
# Latihan Unpacking
data_user = ["Budi", 25, "Jakarta"]
nama, umur, kota = data_user

print(f"User {nama} tinggal di {kota} dan berumur {umur} tahun.")
```

## B. Global Keyword & Scope
Variabel di dalam `def` (fungsi) bersifat **Local**. Gunakan keyword `global` jika ingin memodifikasi variabel di luar fungsi.

```python
poin = 0

def tambah_poin():
    global poin
    poin = poin + 10

tambah_poin()
print("Poin sekarang:", poin) # Hasilnya harus 10
```

## C. String Indexing & Slicing
Format: `string[start : stop : step]`
- `start`: Index awal (ikut diambil).
- `stop`: Index akhir (TIDAK ikut diambil).
- `step`: Jarak lompatan antar karakter.

```python
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Ambil karakter ke-10 (Index 9)
print("Index 9:", alphabet[9])

# Ambil potongan dari index 0 sampai 5 (ABCDE)
print("Slicing 0-5:", alphabet[0:5])

# Balikkan urutan string (Reverse)
print("Reverse:", alphabet[::-1])
```
