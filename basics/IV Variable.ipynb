# Variables Naming

### 1. 🏷️ Konsep Variabel di Python

Di Python, variabel itu bukan "kotak" yang nyimpen nilai, tapi **"label"** yang nempel ke objek di memori.

---

### ✍️ Multi-Word Names
Karena kita tidak bisa menggunakan spasi dalam penamaan variabel, ada 3 gaya penulisan yang umum digunakan:

*   **Camel Case**: `myVariableName` (Huruf pertama kecil, kata selanjutnya kapital).
*   **Pascal Case**: `MyVariableName` (Setiap awal kata menggunakan huruf kapital).
*   **Snake Case**: `my_variable_name` (Menggunakan underscore sebagai pemisah).

> **Note:** **Snake Case** adalah standar resmi Python berdasarkan panduan **PEP 8**.

````python
myVariableName = "John" # Camel Case
my_variable_name = "John ini yang paling di rekomendasikan" # Snake Case
MyVariableName = "John" # Pascal Case
myvariablename = "John" # Flat Case
print(myVariableName) # Hasilnya: John
print(my_variable_name) # Hasilnya: John
print(MyVariableName) # Hasilnya: John
print(myvariablename) # Hasilnya: John
````

## 2. 📜 Rules for Python Variables

*   **Karakter Pertama:** 
    *   Nama variabel **harus** dimulai dengan huruf atau karakter underscore (`_`).
    *   Nama variabel **tidak boleh** dimulai dengan angka.
*   **Karakter Pendukung:** 
    *   Hanya boleh mengandung karakter alfanumerik dan underscore (`A-z`, `0-9`, dan `_`).
*   **Case-Sensitive:** 
    *   Variabel bersifat peka terhadap huruf besar/kecil. Contohnya: `age`, `Age`, dan `AGE` adalah tiga variabel yang berbeda.
*   **Reserved Keywords:** 
    *   Nama variabel **tidak boleh** menggunakan kata kunci bawaan Python (seperti `if`, `else`, `while`, `import`, dll).

---

## 3.  ✨ Fitur "Sakti" Python: Multiple Assignment

Ini fitur Python yang sering bikin bingung pemula tapi sangat berguna untuk efisiensi kode:

* **Many Values to Multiple Variables**:
    * **Lo bisa isi banyak variabel sekaligus dalam satu baris**:
    * `x, y, z = "Orange", "Banana", "Cherry"`

> **Penting:** Pastikan jumlah variabel di sisi kiri **sama** dengan jumlah nilai di sisi kanan, kalau nggak Python bakal kasih `ValueError`.

````python
x,y,z = "aku", "kamu", "dia"
print(x)
print(y)
print(z)
````

* **One Value to Multiple Variables**:
    * **Satu nilai yang sama diberikan ke banyak variabel sekaligus**:
    * `x = y = z = "Orange"`
    * Di sini, variabel `x`, `y`, dan `z` semuanya merujuk ke satu objek yang sama yaitu `"Orange"`.

````python
x = y = z = "aku"
print(x)
print(y)
print(z)
````

### 📦 Unpack a Collection

Jika kamu memiliki sekumpulan nilai dalam sebuah **List** atau **Tuple**, Python memungkinkan kamu untuk mengekstrak nilai-nilai tersebut ke dalam variabel-variabel terpisah.

* **Cara Unpacking**:
    * `fruits = ["apple", "banana", "cherry"]`
    * `x, y, z = fruits`

* **Hasilnya**:
    * `x` akan berisi `"apple"`
    * `y` akan berisi `"banana"`
    * `z` akan berisi `"cherry"`

> **Catatan:** Jumlah variabel di sisi kiri harus **sama persis** dengan jumlah item yang ada di dalam koleksi tersebut, jika tidak maka akan muncul `ValueError`.

````python
# Latihan Unpacking
data_user = ["Budi", 25, "Jakarta"]
nama, umur, kota = data_user

print(f"User {nama} tinggal di {kota} dan berumur {umur} tahun.")
````

## 4.  📤 Output Variables

Fungsi `print()` adalah cara paling umum untuk menampilkan output di Python. Ada beberapa cara untuk menggabungkan variabel dan string:

1.  **Menggunakan Koma (`,`):**
    *   Bisa menggabungkan tipe data berbeda (misal: string + angka).
    *   Otomatis nambahin spasi di antara item.
    *   `x = "Python"; y = "seru"`
    *   `print(x, y)` -> Hasil: `Python seru`

2.  **Menggunakan Operator Plus (`+`):**
    *   Hanya bisa untuk tipe data yang **sama** (misal: string + string).
    *   Tidak otomatis nambahin spasi.
    *   `print(x + y)` -> Hasil: `Pythonseru`
    *   ⚠️ **Bahaya:** `print("Umur saya " + 25)` bakal **Error** karena string nggak bisa ditambah angka secara langsung.

3.  **Menggunakan f-string (Cara Modern):**
    *   Paling fleksibel dan bersih.
    *   `age = 25`
    *   `print(f"Umur saya {age}")`

````python
# 1 Menggunakan Comma

x = "Aldi" #Aldis burger burgernya lembut dagingnya
y = "Taher" #Juicy Luicy Rizky Febian Mahalini
print(x,y)
````

````python
# 2 Menggunakan Plus
x = "Hello World"
y = "ArithmeticError"
print( y + " " + x)
````

````python
# 3 Menggunakan Format String
x = "Hello World"
y = "Arithmetic Error"
print (f" Hallo kebanyakan {y} bikin error {x}")
````

### 🧠 Evaluasi: Bedanya Koma (`,`) vs Plus (`+`) vs f-string

Banyak yang kegocek di bagian spasi dan tipe data. Ini rangkuman aturannya:

#### 1. Menggunakan Koma (`,`)
*   **Spasi Otomatis**: Python otomatis menyelipkan satu spasi di setiap koma.
*   **Beda Tipe Data**: Bisa menggabungkan String dengan Integer/Float tanpa error.
*   *Contoh:* `print("Skor:", 100)` -> `Skor: 100`

#### 2. Menggunakan Plus (`+`)
*   **Tanpa Spasi**: Menempelkan string apa adanya. Kalau mau spasi, harus tulis manual `" "`.
*   **Wajib String**: Bakal **Error** kalau coba gabungin String + Angka.
*   *Contoh:* `print("Skor:" + "100")` -> `Skor:100`

#### 3. Menggunakan f-string (`f""`)
*   **Presisi**: Lo yang nentuin posisi spasi di dalam string.
*   **Auto-convert**: Otomatis mengubah angka di dalam `{}` jadi teks.
*   *Contoh:* `score = 100; print(f"Skor: {score}")` -> `Skor: 100`

> **💡 Tips:** Kalau mau aman dan rapi, selalu pakai **f-string**. Kalau mau cepat buat debugging, pakai **koma**. Hindari `+` kecuali lo emang mau nempelin string tanpa jarak.

## B. Global Keyword & Scope
Variabel di dalam `def` (fungsi) bersifat **Local**. Gunakan keyword `global` jika ingin memodifikasi variabel di luar fungsi.

### 🌐 Global Variables

Variabel yang dibuat di luar fungsi (seperti yang kita lakuin dari tadi) disebut **Global Variables**. Variabel ini bisa dipakai oleh siapa saja, baik di dalam maupun di luar fungsi.

#### 1. Contoh Variabel Global
```python
x = "awesome" # Ini variabel global

def myfunc():
  print("Python is " + x) # Bisa diakses dari dalam fungsi

myfunc()

````python
x = "Luar Biasa"

def Function():
    print("Python itu " + x)
Function()

````

````python
poin = 0

def tambah_poin():
    global poin
    poin = poin + 10

tambah_poin()
print("Poin sekarang:", poin) # Hasilnya harus 10
````

### 🛠️ Memahami `def` (Function) & Global Variable

Banyak yang bingung kenapa harus pakai `def`. Anggap saja `def` itu cara kita membuat **"Tombol Perintah"**.

#### 1. Cara Kerja Fungsi
*   **Mendefinisikan (`def`)**: Kita merakit tombolnya dan kasih nama. Di tahap ini, perintah **belum dijalankan**.
*   **Memanggil (`Function()`)**: Kita menekan tombolnya. Di sinilah perintah di dalamnya **dijalankan**.

```python
x = "Luar Biasa" # Variabel Global (Bisa diakses siapa saja)

def Sapa():
    # Ini cuma instruksi, belum jalan
    print("Halo!")
    print("Python itu " + x)

# --- PROGRAM UTAMA ---
print("Mulai...") 
Sapa() # Nah, baru di sini "tombol" Sapa ditekan

````python
x = "Luar Biasa"

def Sapa():
    print("Halo!")

# Gue nggak panggil Sapa() di sini.
print("Python " + x)
Sapa()
````

### 🏘️ Analogi Real Life: Fungsi vs Variabel

| Komponen | Nama di Kode | Analogi Dunia Nyata | Peran |
| :--- | :--- | :--- | :--- |
| **Global Variable** | `x = "Luar Biasa"` | **Papan Pengumuman** | Informasi umum yang bisa dibaca siapa saja di seluruh kantor. |
| **Function Definition** | `def Sapa():` | **SOP / Prosedur** | Buku panduan yang nulis: "Kalau ada tamu, ucapkan Halo!". |
| **Function Body** | `print("Halo!")` | **Isi Perintah** | Tindakan nyata yang dilakukan (teriak "Halo"). |
| **Function Call** | `Sapa()` | **Eksekusi** | Bos nyuruh staf: "Kerjakan SOP Sapa sekarang!" |

### 👥 Si Kembar yang Berbeda (Local vs Global)

Pahami perbedaan ini supaya lo nggak bingung kenapa nilainya berubah atau tetap:

#### 1. Kasus "Dua Orang Berbeda" (Tanpa Keyword Global)
Di sini ada dua `x`. `x` yang di luar dan `x` yang di dalam fungsi. Mereka cuma punya nama yang sama, tapi dompetnya beda.
```python
x = "Luar"

def Fungsi():
    x = "Dalam" # Python bikin x BARU (Lokal)
    print(x)    # Output: Dalam

Fungsi()
print(x)        # Output: Luar (Karena x yang di luar nggak diganggu)

````python
x = "Luar"

def Fungsi():
    global x    # "Jangan bikin baru, pake x yang di luar!"
    x = "Dalam" # Ngerubah x yang sama

Fungsi()
print(x)        # Output: Dalam (Nilai di luar IKUT BERUBAH)
````
