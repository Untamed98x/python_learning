# 📑 Lanjutan List Mastery: Modifikasi Data

Karena List bersifat **Mutable**, kita bisa bebas "mengacak-acak" isinya tanpa harus bikin variabel baru.

### 1. Menambah Data Lebih Spesifik
Selain `.append()`, ada cara lain buat nambahin data:

* **`.insert(index, data)`**: Nyelipin data di posisi tertentu (index).
* **`.extend(list_lain)`**: Gabungin dua list jadi satu (bukan bikin list di dalam list).

```python
# Contoh Nyata
keranjang = ["Kopi", "Gula"]

# Selipin "Susu" di posisi paling depan (index 0)
keranjang.insert(0, "Susu") 
# Hasil: ["Susu", "Kopi", "Gula"]

# Tambah banyak barang sekaligus
tambahan = ["Roti", "Selai"]
keranjang.extend(tambahan)
# Hasil: ["Susu", "Kopi", "Gula", "Roti", "Selai"]

````python
keranjang = ["kopi","Susu"]
keranjang.insert(3, "Gula") # ✅ BENAR: "Keranjang, taruh gula di depan kopi!"
print(keranjang) # Output: ['Gula', 'kopi', 'Susu
````

````python
tambahan = ["selai", "mentega", "roti"]
keranjang.extend(tambahan) # ✅ BENAR: "Keranjang, tambahin selai, mentega, dan roti di belakang gula!"
print(keranjang) # Output: ['Gula', 'kopi', 'Susu', 'selai', 'mentega', 'roti']
````

# 🗿 Tuple Mastery: Koleksi Data "Segel Mati"

Tuple adalah tipe data koleksi yang bersifat **Immutable**. Artinya, setelah dibuat, isinya tidak bisa ditambah, dikurangi, atau diubah.

### 1. Kenapa Pakai Tuple?
*   **Keamanan**: Data tidak sengaja terhapus/berubah oleh kode lain.
*   **Performa**: Lebih cepat diproses oleh Python dibanding List.
*   **Konstanta**: Cocok untuk data permanen (Contoh: Nama hari, Koordinat GPS, Settingan Admin).

### 2. Operasi yang Tersedia
Hanya ada dua method utama pada Tuple:
*   `.count(item)` : Menghitung jumlah kemunculan data.
*   `.index(item)` : Mencari nomor index dari data tersebut.

### 3. Teknik Unpacking
Cara tercepat memindahkan isi Tuple ke variabel terpisah.
```python

emas, perak, perunggu = pemenang

````python
Data_Tuple = (1, 2, 3, 4, 5) # tipe data tuple yang berisi angka
print(Data_Tuple)
````

### 🛠️ Klasifikasi Method Python: Aksi vs Informasi

Jangan tertukar saat menggunakan fungsi di dalam `print()`.

| Jenis | Contoh Method | Cara Penggunaan | Hasil Jika di-Print |
| :--- | :--- | :--- | :--- |
| **AKSI** (In-place) | `.append()`, `.insert()`, `.remove()`, `.sort()`, `.reverse()` | Jalankan dulu perintahnya, baru panggil variabelnya di baris baru. | `None` (Kosong) |
| **INFO** (Return) | `.count()`, `.index()`, `len()`, `type()` | Bisa langsung dimasukkan ke dalam `print()`. | Nilai/Data asli |

### 🔍 Logika `.count(x)` : Pencarian Spesifik

Ingat! `.count(x)` **BUKAN** menghitung urutan angka 1 sampai `x`.

*   **Input**: `x` adalah benda/nilai yang ingin lo cari.
*   **Proses**: Python akan keliling di dalam Tuple/List, lalu setiap nemu benda yang **identik** dengan `x`, dia bakal tambah poin 1.
*   **Output**: Total poin (berapa banyak benda itu ditemukan).

#### 🛠️ Contoh Salah Kaprah:
```python
data = (5, 5, 5)

# Kalau lo kira ini ngitung 1, 2, 3, 4, 5... LO SALAH.
# Ini artinya: "Ada berapa angka 5 di sini?"
print(data.count(5)) # Hasil: 3

````python
print(Data_Tuple.count(1))
````

### 🛑 Stop! Kenapa `.insert()` Gagal di Tuple?

Jangan tertukar antara **List** dan **Tuple**. Keduanya punya aturan main yang sangat berbeda.

| Fitur | List `[ ]` (Mutable) | Tuple `( )` (Immutable) |
| :--- | :--- | :--- |
| **Tambah Data** (`.append`) | ✅ BISA | ❌ GAGAL (Error) |
| **Selippin Data** (`.insert`) | ✅ BISA | ❌ GAGAL (Error) |
| **Hapus Data** (`.remove`) | ✅ BISA | ❌ GAGAL (Error) |
| **Ubah Isi** (`data[0] = x`) | ✅ BISA | ❌ GAGAL (Error) |
| **Keamanan Data** | Rendah (Bisa berubah) | **Tinggi (Permanen)** |

#### 💡 Solusi Jika Mau Mengubah Tuple:
Kalau lo terpaksa banget mau nambahin data ke Tuple, lo harus "reinkarnasiin" dulu ke List:
1. `temp = list(Data_Tuple)` (Ubah ke List)
2. `temp.insert(0, 6)` (Insert datanya)
3. `Data_Tuple = tuple(temp)` (Balikin lagi ke Tuple)

````python
Data_Tuple.insert(0,6)
print(Data_Tuple)
````


#### 💡 Golden Rule:
Jika method tersebut **mengubah** isi data asli (seperti menambah/menghapus), jangan dimasukkan ke dalam `print()`. Cukup jalankan perintahnya saja.
