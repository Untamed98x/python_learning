Di Python, tipe data menentukan operasi apa yang bisa dilakukan pada suatu variabel. Berikut adalah daftar lengkapnya:

---

### 1. Tipe Data Dasar (Primitive)
Ini adalah tipe data paling sederhana yang menyimpan satu nilai.

| Tipe Data | Nama (Class) | Contoh | Deskripsi |
| :--- | :--- | :--- | :--- |
| **String** | `str` | `"Ojay"`, `'123'` | Teks/Karakter (Wajib pakai tanda petik). |
| **Integer** | `int` | `10`, `-5`, `1000` | Angka bulat tanpa desimal. |
| **Float** | `float` | `3.14`, `10.0`, `2.5` | Angka desimal (pakai titik, bukan koma). |
| **Boolean** | `bool` | `True`, `False` | Nilai logika (Hanya Benar atau Salah). |

---



````python
Primitive_Str= "Ojay", "123"
Primitive_Int= 10,-5,1000
Primitive_Float= 3.14, -0.001, 2.71828
````

### 2. Tipe Data Koleksi (Non-Primitive)
Digunakan untuk menyimpan banyak data dalam satu variabel.

| Tipe Data | Nama (Class) | Contoh | Karakteristik |
| :--- | :--- | :--- | :--- |
| **List** | `list` | `["a", "b", 1]` | Berurutan, bisa diubah (**Mutable**), pakai `[]`. |
| **Tuple** | `tuple` | `(1, 2, 3)` | Berurutan, GAK bisa diubah (**Immutable**), pakai `()`. |
| **Range** | `range` | `range(6)` | Urutan angka otomatis (sering buat looping). |
| **Dictionary** | `dict` | `{"id": 1}` | Pasangan **Key: Value**, pakai `{}`. |
| **Set** | `set` | `{"apel", "jeruk"}` | Tidak berurutan, unik (nggak boleh ada duplikat). |
