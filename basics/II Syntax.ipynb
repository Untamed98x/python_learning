## 3. Statements

Dalam bahasa pemrograman Python, secara sederhana statement adalah satu baris perintah atau instruksi yang bisa langsung dijalankan oleh komputer. Kalau diibaratkan kalimat dalam bahasa manusia, statement adalah "kalimat perintah" yang menyuruh Python melakukan sesuatu.

1. Statement Satu Baris

Biasanya, setiap satu baris kode di Python dianggap sebagai satu statement. Python tidak mewajibkan penggunaan titik koma (;) di akhir baris seperti di bahasa C++ atau Java.

```python
print("Halo dunia!")  # Ini satu statement untuk mencetak teks
x = 10               # Ini satu statement untuk menyimpan angka
```

2. Multi-line Statement (Statement yang diputus)

Kadang, satu perintah bisa sangat panjang. Kamu bisa memutusnya menjadi beberapa baris menggunakan tanda backslash (\) agar kode lebih rapi dan mudah dibaca.

1. Tombol Rahasia

Pada layout Jepang, simbol backslash biasanya berbagi tempat dengan simbol Yen (¥).

Tekan tombol Option (⌥) + ¥.

```python
total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)
```

```python
print("Hello World!")
print("Have a good day.")
print("Learning Python is fun!")
```

4. Statement Majemuk (Compound Statements)

Ini adalah statement yang mengandung statement lain di dalamnya, contohnya seperti if, while, atau for. Mereka biasanya diakhiri dengan tanda titik dua (:) dan baris berikutnya harus menjorok ke dalam (indentasi).

```python
if x > 5:
    print("X besar")  # Statement ini ada di dalam blok 'if'
```
