# 🗿 Tuple Mastery: Koleksi Data "Segel Mati"

### 🛠️ Klasifikasi Method Python: Aksi vs Informasi

*   **Output**: Total poin (berapa banyak benda itu ditemukan).

### 🛑 Stop! Kenapa `.insert()` Gagal di Tuple?

Jangan tertukar antara **List** dan **Tuple**. Keduanya punya aturan main yang sangat berbeda.

| Fitur | List `[ ]` (Mutable) | Tuple `( )` (Immutable) |
| :--- | :--- | :--- |
| **Tambah Data** (`.append`) | ✅ BISA | ❌ GAGAL (Error) |
| **Selippin Data** (`.insert`) | ✅ BISA | ❌ GAGAL (Error) |
| **Hapus Data** (`.remove`) | ✅ BISA | ❌ GAGAL (Error) |
| **Ubah Isi** (`data[0] = x`) | ✅ BISA | ❌ GAGAL (Error) |
| **Keamanan Data** | Rendah (Bisa berubah) | **Tinggi (Permanen)** |

#### 💡 Solusi Jika Mau Mengubah Tuple:
Kalau lo terpaksa banget mau nambahin data ke Tuple, lo harus "reinkarnasiin" dulu ke List:
1. `temp = list(Data_Tuple)` (Ubah ke List)
2. `temp.insert(0, 6)` (Insert datanya)
3. `Data_Tuple = tuple(temp)` (Balikin lagi ke Tuple)

````python
Data_Tuple.insert(0,6)
print(Data_Tuple)
````


#### 💡 Golden Rule:
Jika method tersebut **mengubah** isi data asli (seperti menambah/menghapus), jangan dimasukkan ke dalam `print()`. Cukup jalankan perintahnya saja.

#### 1. Operasi Himpunan
Set di Python itu jagonya matematika himpunan. Bayangin lo punya dua kelompok user: Tim Python dan Tim Java.


````python
# Data himpunan
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Hitung operasi himpunan
union = A | B
intersection = A & B
difference_ab = A - B
difference_ba = B - A
symmetric_difference = A ^ B

# Buat figure 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Visualisasi Operasi Himpunan', fontsize=16)

# 1. Union
venn = venn2([A, B], set_labels=('A', 'B'), ax=axs[0, 0])
venn.get_patch_by_id('10').set_color('skyblue')
venn.get_patch_by_id('01').set_color('lightgreen')
venn.get_patch_by_id('11').set_color('orange')
axs[0, 0].set_title(f'Union\nA ∪ B = {union}')

# 2. Intersection
venn = venn2([A, B], set_labels=('A', 'B'), ax=axs[0, 1])
venn.get_patch_by_id('11').set_color('red')
venn.get_patch_by_id('10').set_alpha(0.2)
venn.get_patch_by_id('01').set_alpha(0.2)
axs[0, 1].set_title(f'Intersection\nA ∩ B = {intersection}')

# 3. Difference A - B
venn = venn2([A, B], set_labels=('A', 'B'), ax=axs[1, 0])
venn.get_patch_by_id('10').set_color('purple')
venn.get_patch_by_id('01').set_alpha(0.2)
venn.get_patch_by_id('11').set_alpha(0.2)
axs[1, 0].set_title(f'Difference\nA - B = {difference_ab}')

# 4. Symmetric Difference
venn = venn2([A, B], set_labels=('A', 'B'), ax=axs[1, 1])
venn.get_patch_by_id('10').set_color('gold')
venn.get_patch_by_id('01').set_color('gold')
venn.get_patch_by_id('11').set_alpha(0.1)
axs[1, 1].set_title(f'Symmetric Difference\nA △ B = {symmetric_difference}')

plt.tight_layout()
plt.show()

# Print hasil operasi
print('A =', A)
print('B =', B)
print('Union:', union)
print('Intersection:', intersection)
print('A - B:', difference_ab)
print('B - A:', difference_ba)
print('Symmetric Difference:', symmetric_difference)
````

#### A. Intersection (Irisan)

Nyari siapa aja yang ada di kedua kelompok.

````python
python_devs = {"Ojay", "Budi", "Ani"}
java_devs = {"Budi", "Siti", "Rudi"}

# Siapa yang jago dua-duanya?
dua_duanya = python_devs.intersection(java_devs)
# Atau pake simbol: python_devs & java_devs

print(dua_duanya) # Output: {'Budi'}
````

#### B. Union (Gabungan)

Nggabungin semua orang jadi satu kelompok besar (tanpa duplikat).

````python
semua_devs = python_devs.union(java_devs)
# Atau pake simbol: python_devs | java_devs

print(semua_devs) # Output: {'Ojay', 'Budi', 'Ani', 'Siti', 'Rudi'}
````

#### C. Difference (Selisih)

Nyari siapa yang ada di Tim A tapi nggak ada di Tim B.

````python
semua_devs = python_devs.difference(java_devs)
print(semua_devs) # Output: {'Ojay', 'Ani'}
````

#### D. Symmetric Difference (^)

Kalau Difference itu nyari yang ada di A tapi GAK ADA di B, nah Symmetric Difference itu nyari yang ada di A atau B, tapi BUKAN di keduanya.

Gampangnya: "Siapa aja yang cuma jago SATU bahasa (Python doang atau Java doang)?"

````python
python_devs = {"Ojay", "Budi"}
java_devs = {"Budi", "Siti"}

# Pake simbol topi/caret (^)
cuma_satu_keahlian = python_devs ^ java_devs

print(cuma_satu_keahlian) # Output: {'Ojay', 'Siti'}
# Budi ilang karena dia ada di dua-duanya (irisan).
````


## C. Frozenset
* **Fungsi**: `frozenset({ 1,2,3})`
* **Sifat**: `Permanen(Immutable)`
```python
data_tetap = frozenset({1, 2, 3})
# data_tetap.add(4) # Ini bakal ERROR
```


#### 1. Kenapa Harus "Dibekukan"?

* **Di Python**, ada aturan main: Key dalam Dictionary harus bersifat Immutable (tidak bisa berubah).
* Set biasa itu mutable `(bisa .add()` atau `.remove())`.
* Jadi, Set nggak bisa jadi Key di Dict.

Frozenset itu immutable. Jadi, dia bisa jadi Key.

##### 2. Cara Membuat Frozenset

Lo nggak bisa pake kurung kurawal {} langsung. Lo harus pakai fungsi frozenset().


````python
# Set biasa
hobi_ojay = {"Coding", "Gaming"}
hobi_ojay.add("Tidur") # Bisa banget!

# Frozenset
hobi_beku = frozenset(["Coding", "Gaming"])
# hobi_beku.add("Tidur") # ERROR! Python bakal protes karena frozenset itu beku, gak bisa diubah setelah dibuat.

print(hobi_ojay)
print(hobi_beku)
````

#### 3. Kekuatan Frozenset di Dictionary

* Ini skenario paling berguna buat Frozenset.
* Bayangin lo mau bikin kategori berdasarkan kumpulan skill:

````python
# Kita punya kumpulan skill sebagai Frozenset
skill_ai = frozenset(["Python", "Math", "Statistics"])
skill_web = frozenset(["HTML", "CSS", "JS"])

# Gunakan Frozenset sebagai KEY di Dictionary
gaji_standar = {
    skill_ai: 20_000_000,
    skill_web: 15_000_000
}

print(f"Gaji AI: {gaji_standar[skill_ai]}")
````

| **Generator** | `(x for x in r)` | Koleksi data hemat memori (Lazy Loading). |
| **Slice** | `slice(0, 5)` | Objek untuk memotong urutan (List/String). |
