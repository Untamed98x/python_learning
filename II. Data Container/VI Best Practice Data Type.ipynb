

## 2. String Manipulation: The "Slicing" Magic
Slicing adalah cara motong teks. Rumusnya: `[start : stop : step]`



*   **Slicing Dasar**:
    ```python
    nama = "Python"
    print(nama[0:2])  # "Py" (Ambil indeks 0 sampai sebelum 2)
    print(nama[2:])   # "thon" (Ambil dari indeks 2 sampai habis)
    print(nama[-1])   # "n" (Ambil karakter paling terakhir)
    ```
*   **Trick Kebalik**: Cara paling cepet buat balikkin tulisan.
    ```python
    pesan = "makaN"
    print(pesan[::-1]) # "Nakam"
    ```


## 3. String Best Practices (Cleaning Data)
Data dari luar biasanya kotor (banyak spasi liar atau huruf besar-kecil gak jelas).

*   **Strip**: Buang spasi di ujung-ujung.
    ```python
    user = "  ojay   "
    print(user.strip()) # "ojay"
    ```
*   **Replace**: Ganti kata tertentu.
    ```python
    txt = "Saya suka Java"
    print(txt.replace("Java", "Python")) # "Saya suka Python"
    ```
*   **Split & Join**: Pecah teks jadi List, atau gabungin List jadi Teks.
    ```python
    hobi = "coding,gaming,tidur"
    list_hobi = hobi.split(",") # Jadi ['coding', 'gaming', 'tidur']
    ```

---

````python
kode = "##DATA_KABUPATEN_1##"
print(kode.strip("#&1"))
````

````python
print(kode.lstrip("#"))
print(kode.rstrip("#"))
````

````python
text = "www.googlecom.com"

# Lo mau buang 'www.' dan '.com'
print(text.strip("w.com"))
# Hasil: "google"
````

````python
# Pecah berdasarkan titik
parts = text.split(".")

print(parts) # Hasilnya List: ['www', 'googlecom', 'com']

# Ambil elemen tengah (indeks 1)
domain = parts[1]
print(domain) # Hasil: "googlecom"
````

````python
#Method Chaining

web = "www.fauzy.com.com"
clean_web = web.removeprefix("www.").removesuffix([".com",".id",".org"])

print(clean_web) # Hasil: fauzy.com
````

Sesuai namanya, Chain itu artinya Rantai. Jadi, Method Chaining adalah teknik menyambung (merangkai) beberapa method sekaligus dalam satu baris kode, dengan memanggilnya berderet pakai titik.

Ibaratnya lo lagi main game fighting, ini tuh Combo Skill. Alih-alih jalanin jurus satu per satu dan nunggu, lo langsung masukin urutan jurus sekaligus.

1. Analogi Pabrik

Bayangin lo punya bahan baku berupa "Ayam Mentah".
Kalau pakai cara biasa, prosesnya begini:

Ayam dicuci → jadi Ayam Bersih.

Ayam Bersih dipotong → jadi Ayam Potong.

Ayam Potong digoreng → jadi Ayam Goreng.

Nah, pakai Method Chaining, instruksinya dipersingkat jadi satu jalur mesin conveyor belt:
Ayam.cuci().potong().goreng() → Langsung keluar jadi Ayam Goreng!

Perbandingan Kode (Biasa vs Chaining)

Misal lo dapet data berantakan dari form input user: "   oJaY gAnTeng   ".
Tujuan lo: Buang spasi berlebih, kecilin semua hurufnya, lalu bikin huruf depannya aja yang kapital.

Cara Lama (Tanpa Chaining):

````python
data_kotor = "   oJaY gAnTeng   "

# Lo harus bikin variabel sementara berkali-kali
step_1 = data_kotor.strip()   # Hasil: "oJaY gAnTeng"
step_2 = step_1.lower()       # Hasil: "ojay ganteng"
hasil_akhir = step_2.title()  # Hasil: "Ojay Ganteng"

print(hasil_akhir)
````

````python
data_kotor = "   oJaY gAnTeng   "

# Langsung sambung semua method dalam 1 baris
hasil_akhir = data_kotor.strip().lower().title()

print(hasil_akhir)
````
