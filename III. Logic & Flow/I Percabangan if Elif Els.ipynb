## 1. Percabangan: if, elif, else
Definisi: Cara program mengambil keputusan berdasarkan kondisi Boolean (True atau False).

* if: Cek kondisi utama.

* elif: Singkatan dari else if. Dipakai kalau ada banyak opsi (misal: menu kalkulator).

* else: "Tong sampah" terakhir kalau semua kondisi di atas nggak ada yang bener.

```⚠️ Inget: Pake == untuk membandingkan, bukan = (untuk isi variabel). Dan jangan lupa Indentasi (spasi masuk)!```

Percabangan (`if`, `elif`, `else`) adalah cara kita memberikan "insting" kepada Python untuk mengambil keputusan. Tanpa ini, kode lo hanyalah baris perintah mati yang berjalan lurus ke bawah.

## 1. Definisi & Aturan Main

### A. Anatomi Dasar
Setiap blok percabangan wajib mengikuti aturan ini:
1. **Kondisi**: Sesuatu yang menghasilkan `True` atau `False`.
2. **Titik Dua (`:`)**: Penanda bahwa blok instruksi akan dimulai.
3. **Indentasi (Spasi)**: Python menggunakan spasi (biasanya 4 spasi atau 1 Tab) untuk menentukan kode mana yang merupakan "anak" dari perintah `if`.

http://googleusercontent.com/image_content/134



### B. Keywords
*   **`if`**: Pintu pertama. Wajib ada.
*   **`elif` (else if)**: Pintu tambahan jika pintu pertama tertutup. Bisa ada banyak `elif`.
*   **`else`**: Pintu darurat jika semua pintu di atas tertutup. Tidak butuh kondisi lagi.

---

## 2. Best Practices & Common Mistakes

| Do's ✅ | Don'ts ❌ |
| :--- | :--- |
| Gunakan `==` untuk membandingkan nilai. | Jangan gunakan `=` (ini untuk mengisi variabel). |
| Selalu lakukan *Data Cleaning* sebelum `if`. | Jangan cek input mentah (raw input) karena sensitif huruf besar/kecil. |
| Gunakan `in` untuk cek keberadaan di List/Set. | Jangan gunakan banyak `if` terpisah jika kondisinya saling berkaitan. |

# 2. Gerbang Bertingkat : ELIF

## 1. Struktur Dasar: Cabang Logika

```if: Gerbang pertama. Kalau syaratnya tembus, masuk sini.```

* ```elif (singkatan dari else if): Gerbang cadangan. Kalau gerbang pertama Gagal, Python bakal ngecek gerbang ini.```
*  ```Lo bisa punya banyak elif.```
* ```else: Pintu darurat terakhir. Kalau SEMUA gerbang di atas gagal, otomatis masuk sini.```*

````python
Customer = {"nama": "Budi", "total_belanja": 150000, "kode_promo": "DISKON50"}

# KONDISI 1 (IF)
if Customer["total_belanja"] > 200000:
    print("Mendapatkan Voucher Hotel!")

# KONDISI 2 (ELIF) - Dicek cuma kalau Kondisi 1 FALSE
elif Customer["total_belanja"] > 100000:
    print("Mendapatkan Voucher Makan!")

# KONDISI 3 (ELSE) - Jalan kalau semua di atas FALSE
else:
    print("Terima kasih sudah belanja, yuk tambah lagi!")
````
