## 1. Percabangan: if, elif, else
Definisi: Cara program mengambil keputusan berdasarkan kondisi Boolean (True atau False).

* if: Cek kondisi utama.

* elif: Singkatan dari else if. Dipakai kalau ada banyak opsi (misal: menu kalkulator).

* else: "Tong sampah" terakhir kalau semua kondisi di atas nggak ada yang bener.

```⚠️ Inget: Pake == untuk membandingkan, bukan = (untuk isi variabel). Dan jangan lupa Indentasi (spasi masuk)!```

# Kalau variabel adalah wadah, maka percabangan ini adalah papan penunjuk jalan. Bayangkan kamu sedang mengemudi:

* if: Kamu ketemu pertigaan pertama. "Kalau ke kiri arah Jakarta, belok kiri."
* elif: Kalau ternyata jalan ke kiri ditutup, kamu liat papan kedua. "Kalau ke kanan arah Bandung, belok kanan."

* else: Kalau kedua jalan di atas nggak bisa dilewati, kamu terpaksa putar balik (pilihan terakhir).

### 📝 Latihan Mandiri: Basic If Statement

#### 🌡️ Analogi: Protokol Keamanan Server
Bayangkan kamu punya sensor suhu yang terus-menerus mengirim data angka ke komputer. Kamu butuh sistem yang bereaksi beda-beda tergantung angka tersebut.

#### Skenario: "Sistem Sensor Suhu Ruangan"
Kalau suhu di atas 35 derajat, munculin peringatan.



````python
# Variabel
# --- 1. SET INPUT ---
suhu = 38

# --- 2. LOGIC ENGINE (Satu Kesatuan) ---
if suhu > 35:
    # Blok ini buat kondisi paling ekstrim
    status = "🔴 KRITIS"
    aksi = "Matikan Server!"

elif suhu >= 30:
    # Blok ini saringan kedua (hanya jika suhu <= 35)
    status = "🟡 WASPADA"
    aksi = "Nyalakan Pendingin Cadangan."

else:
    # Blok ini saringan terakhir (jika semua di atas salah)
    status = "🟢 AMAN"
    aksi = "Operasi Normal."

# --- 3. OUTPUT ---
print(f"Suhu: {suhu}°C | Status: {status} | Action: {aksi}")
````

#### Cell 1: Logika Terpisah (Independent If)
Gunakan ini kalau lo punya dua hal yang nggak nyambung buat dicek. 
Komputer bakal ngecek keduanya, nggak peduli yang satu benar atau salah.


````python
# Cek 1: Status Listrik
listrik_on = True
if listrik_on:
    print("⚡ Listrik Terkoneksi")

# Cek 2: Status Pintu (Tidak ada hubungannya sama listrik)
pintu_terkunci = False
if not pintu_terkunci:
    print("🔓 Pintu Terbuka")
````

#### Cell 2: Logika Bertingkat (Nested If)

Ini adalah gabungan. Lo masuk ke pintu pertama, baru ngecek kondisi di dalamnya. Ini sering banget dipake di sistem sensor.

````python
suhu = 38
sistem_aktif = True

if sistem_aktif: # Cek Utama
    print("🖥️ Sistem Monitor Aktif...")

    if suhu > 35: # Cek Cabang di dalam Cek Utama
        print("🔴 ALERT: Suhu Panas!")
    else:
        print("🟢 STATUS: Suhu Normal.")
else:
    print("📴 Sistem sedang Offline.")
````

## 🧭 Python Logic Definition Table

| Keyword | Definisi Teknis | Peran dalam Arsitektur |
| :--- | :--- | :--- |
| **`if`** | Evaluasi kondisi pertama. Jika `True`, jalankan blok kode. | **The Gatekeeper**: Pintu utama masuknya logika. |
| **`elif`** | *Else If*. Mengecek kondisi selanjutnya jika kondisi sebelumnya `False`. | **The Filter**: Saringan bertahap untuk variabel yang sama. |
| **`else`** | Blok yang jalan jika semua kondisi `if` dan `elif` bernilai `False`. | **The Safety Net**: Jaring pengaman untuk data tak terduga. |
| **`and`** | Operator yang menghasilkan `True` hanya jika **semua** syarat benar. | **The Strict Chain**: Semua saklar harus ON. |
| **`or`** | Operator yang menghasilkan `True` jika **salah satu** syarat benar. | **The Alternative**: Satu saklar cukup untuk menyalakan sistem. |
| **`not`** | Membalik nilai Boolean (`True` jadi `False`, dan sebaliknya). | **The Inverter**: Pembalik status (misal: `not mati` berarti `hidup`). |

## 📊 Visualisasi Decision Matrix (Besaran)

**Decision Matrix** adalah sebuah arsitektur pemetaan di mana lo mengolah berbagai variabel input untuk menentukan satu output spesifik berdasarkan kombinasi kondisi tersebut.

Dalam Decision Matrix, lo nggak cuma liat "Ya/Tidak" secara terpisah, tapi lo melihat **Irisan (Intersection)** antar kondisi:

### 🧩 Komponen Input:
1. **Input A**: Status Akses User (`True`/`False`)
2. **Input B**: Kecukupan Saldo (`True`/`False`)
3. **Input C**: Jam Operasional Sistem (`True`/`False`)

### 📋 Tabel Matriks Keputusan:

| Akses (A) | Saldo (B) | Operasional (C) | Output / Hasil |
| :---: | :---: | :---: | :--- |
| ✅ | ✅ | ✅ | **Transaksi Berhasil** |
| ✅ | ❌ | ✅ | **Gagal: Saldo Kurang** |
| ❌ | ✅ | ✅ | **Gagal: Akses Ditolak** |
| ✅ | ✅ | ❌ | **Gagal: Sistem Offline** |

### ⚙️ Logic Gates sebagai "Engine"
Di sinilah **Logic Gates (AND, OR, NOT)** berperan sebagai mesin pengolah matriks tersebut. Tanpa Logic Gates, lo bakal nulis puluhan baris `if` yang berulang. Dengan Logic Gates, lo cukup nulis satu aturan main (SOP):

```python
# Logic Gate Implementation
if punya_akses and saldo_cukup and jam_operasional:
    print("✅ Transaksi Berhasil")
elif not punya_akses:
    print("❌ Gagal: Akses Ditolak")
elif not saldo_cukup:
    print("❌ Gagal: Saldo Kurang")
else:
    print("❌ Gagal: Sistem Offline")

## 🏗️ Core Architecture Theories

### 🪆 1. Teori Russian Dolls (The Nested Hierarchy)
Ini adalah teknik **"Gatekeeping"**. Lo nggak bakal ngecek detail kecil kalau kondisi besarnya belum terpenuhi.
* **Karakteristik:** Vertikal & Dependen.
* **Kapan Pakai:** Saat ada variabel yang menjadi **Syarat Mutlak** bagi variabel lain.

**Contoh Code:**
```python
pusat = True
gardu = True
token = 75

if pusat: # Master Gate
    if gardu: # Sub-Gate
        if token > 0: # Detail
            print("Status: Rumah Terang")
```

### ⚡ 2. Teori High-Voltage Filter (The Elif Chain)
Ini adalah teknik **"Prioritas Baris"**. Komputer Python itu malas, dia bakal berhenti di detik dia nemu jawaban **True**.
* **Karakteristik:** Horizontal & Eksklusif.
* **Logic Flow:** Urutkan dari **Value Tertinggi/Paling Ekstrim** di paling atas.

**Contoh Code:**
```python
suhu = 55

if suhu >= 50: # Saringan rapat (Prioritas Utama)
    print("STATUS: BAHAYA! Suhu sangat tinggi.")
elif suhu >= 30: # Saringan sedang
    print("STATUS: WASPADA! Suhu meningkat.")
else: # Saringan sisa
    print("STATUS: AMAN.")
```

### 🔌 3. Teori Circuit Breaker (The Boolean Logic)
Ini teknik **"Shortcut"**. Kita menggabungkan variabel tanpa peduli hierarki, yang penting semua syarat terpenuhi (AND) atau salah satu cukup (OR).
* **Karakteristik:** Flat & Kombinasi.

**Contoh Code:**
```python
# Mirip logika AND di Excel
if pusat == True and gardu == True and token > 0:
    print("Sistem Berjalan")
```
### 🔌 4. Teori Logic Gates (The Decision Matrix)

Teknik **"Condition Compounding"**. Menggabungkan banyak syarat pake () agar rapi.

AND: Semua syarat wajib True.

OR: Salah satu True sudah cukup.

### 🚦 5. Teori Flagging (The Boolean Toggle)

Teknik "Status Accumulator". Memisahkan proses "Pengecekan" dengan "Eksekusi". Status disimpan dulu di variabel (bendera).

### 🛡️ 5. Teori The Safety Net (The Final Else)
Dalam sistem yang solid, `else` bukan cuma pajangan. Ini adalah **Error Handling**.
* **Definisi:** "Kondisi yang Tidak Terduga" (Edge Case).

**Contoh Code:**
```python
if status == "ON":
    print(1)
elif status == "OFF":
    print(0)
else:
    print("ERROR: Status tidak valid (Check Sensor)")
```

## 📑 SOP Logika Percabangan (Standard Operating Procedure)

1.  **Identifikasi Hierarki:** Apakah variabel A mengontrol variabel B? (Gunakan **Nested/Russian Dolls**).
2.  **Urutkan Prioritas:** Untuk rentang angka, apakah yang paling kritis sudah di paling atas? (Gunakan **Elif Chain**).
3.  **Tentukan Master Switch:** Variabel mana yang kalau `False` langsung mematikan seluruh alur? (Ini jadi `if` paling luar).
4.  **Pasang Jaring Pengaman:** Selalu akhiri dengan `else` untuk menangkap data sampah atau error sensor.

## ⚠️ Common Error: The "IFF" Mistake (Logic Leak)
Dalam Python, tidak ada `iff`. Namun, sering terjadi kesalahan "Logic Leak" di mana pengguna memakai banyak `if` terpisah padahal kondisinya berhubungan.

**Error Example:**
```python
# ❌ SALAH (BOCOR)
if suhu > 30:
    print("Panas")
if suhu > 50:
    print("Sangat Panas")
```
*Jika suhu 60, maka yang tercetak adalah **keduanya**. Ini akan mengacaukan sistem data.*

**Correct Example:**
```python
# ✅ BENAR (EKSLUSIF)
if suhu > 50: # Tertinggi dulu!
    print("Sangat Panas")
elif suhu > 30:
    print("Panas")
```
*Jika suhu 60, dia berhenti di "Sangat Panas" dan tidak akan mencetak "Panas". Logic terjaga.*
