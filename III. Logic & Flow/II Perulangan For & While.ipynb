# 2. Perulangan: for & while
* Definisi: Menjalankan baris kode yang sama berulang kali secara otomatis.

* for: Perulangan yang sudah jelas jumlahnya (misal: sebanyak isi List). Sering dipake bareng enumerate() buat dapet nomor urut.

* while: Perulangan "selama". Dia bakal lari terus selama syaratnya masih True. Bahayanya: bisa jadi infinite loop (gak berhenti-berhenti) kalau syaratnya gak pernah berubah jadi False.
