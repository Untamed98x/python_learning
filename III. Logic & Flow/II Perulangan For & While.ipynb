## 🏗️ Core Architecture: The Looping Theories

Dalam Python, ada dua cara buat bikin perulangan, dan masing-masing punya "kepribadian" yang beda total:

### 🎡 1. The Fixed-Trip Theory (`for` loop)
Ini adalah perulangan yang **"Pasti-Pasti Aja"**. Lo tahu dari awal berapa kali lo mau jalan, atau berapa banyak item yang mau lo proses.
* **Karakteristik:** Terukur, Berhingga, dan Statis.
* **Analogi:** Seperti bis jemputan karyawan yang berhenti di setiap halte yang sudah ada di daftar rute. Kalau halte habis, bis berhenti.
* **Kapan Pakai:** Saat lo punya list data (List, Tuple, String) atau rentang angka yang jelas.



---

### 🌪️ 2. The Conditional Cyclone (`while` loop)
Ini adalah perulangan yang **"Tergantung Situasi"**. Lo nggak tahu kapan bakal berhenti, pokoknya selama syarat masih `True`, hajar terus!
* **Karakteristik:** Dinamis, Berbahaya (bisa bikin *infinite loop*), dan Bergantung pada kondisi.
* **Analogi:** Seperti kipas angin otomatis. Dia bakal muter terus **SELAMA** suhu ruangan di atas 30 derajat. Begitu suhu turun, baru dia stop.
* **Kapan Pakai:** Saat lo menunggu input user, menunggu sensor, atau mencari sesuatu yang lo nggak tahu ada di urutan ke berapa.



---

## 📑 Perbandingan Arsitektur (Summary)

| Fitur | **`for` loop** | **`while` loop** |
| :--- | :--- | :--- |
| **Fokus** | **Iterasi** (Mengunjungi setiap item). | **Kondisi** (Berjalan selama syarat terpenuhi). |
| **Kontrol** | Otomatis (Berhenti di item terakhir). | Manual (Lo harus ubah variabel syarat di dalem). |
| **Resiko** | Rendah (Hampir selalu berhenti). | Tinggi (Bisa bikin laptop nge-hang kalau lupa distop). |

---

## 🛠️ The "Universal Controller": Break & Continue

Ini adalah dua tombol darurat yang bisa lo pasang di dalem `for` maupun `while`:

1. **`break` (The Emergency Stop):** Paksa berhenti sekarang juga, keluar dari perulangan, bodo amat sama sisa antrean.
2. **`continue` (The Skip Button):** Lewatin item yang ini aja, langsung loncat ke item berikutnya.

## 📦 1. Apa itu Iterable? (Bukan Cuma Hitungan)

Bener, hitungan itu salah satunya. Tapi secara teknis, **Iterable** adalah **"Segala sesuatu yang bisa lo preteli satu-satu"**. 

Bayangkan sebuah kontainer. Kalau lo bisa ngambil isinya satu demi satu, itu namanya Iterable.

* **List/Array:** `[10, 20, 30]` → Lo ambil 10, lalu 20, lalu 30. (**Iterable**)
* **String:** `"FAUZY"` → Lo ambil 'F', lalu 'A', lalu 'U', dan seterusnya. (**Iterable**)
* **Range:** `range(1, 5)` → Urutan angka 1 sampai 4. (**Iterable**)
* **Integer Tunggal:** `100` → Lo nggak bisa preteli angka 100. Ini cuma satu nilai utuh. (**Bukan Iterable**).

> **Comprehend:** Loop itu kayak kurir. Dia butuh **alamat tujuan** (Iterable) buat nganter paketnya. Kalau alamatnya cuma satu titik tanpa urutan, kurirnya bingung mau muter ke mana.

## 🏗️ 2. Indentasi (The Border Theory)

Di Python, **Indentasi** (spasi di awal baris) bukan cuma buat rapi-rapian. Itu adalah **"Pagar Wilayah"**. Python nggak pake kurung kurawal `{}` kayak bahasa lain, dia cuma liat spasi.

* **Aturan Main:** Semua kode yang menjorok ke dalam (biasanya 4 spasi atau 1 Tab) dianggap sebagai **"Anak Buah"** dari baris yang ada di atasnya.



**Analogi Ruangan:**
```python
if hari == "hujan": # <-- Tuan Rumah
    bawa_payung = True # <-- Di dalam ruangan (Anak buah if)
    pakai_jas_hujan = True # <-- Di dalam ruangan (Anak buah if)
# -----------------------
print("Selesai") # <-- Di luar ruangan (Gak peduli hujan atau nggak, ini pasti jalan)
```

**Kesalahan Fatal:** Kalau lo lupa indentasi, Python bakal teriak `IndentationError`. Kalau lo salah naruh indentasi (ketinggian atau kerendahan), logika lo bakal **bocor** (kode yang harusnya jalan cuma pas hujan, malah jalan terus).

## 🌪️ 3. Infinite Loop (The Loop of Doom)

Lo bener, ini emang nyata dan sering terjadi di `while` loop. Ini terjadi ketika **Syarat Berhenti** nggak pernah tercapai.

**Analogi Kipas Angin Rusak:**
Lo nyuruh kipas muter: *"Muter terus SELAMA suhu > 30"*. 
Tapi lo lupa masang alat buat nurunin suhu di dalem perulangannya. Hasilnya? Suhu tetep 35, dan kipas muter selamanya sampai laptop lo panas atau baterai abis.



**Contoh Code yang Bahaya:**
```python
baterai = 100
while baterai > 0:
    print("Main Game...")
    # LUPA NULIS: baterai = baterai - 1
    # Hasilnya: baterai tetep 100 selamanya. INFINITE LOOP!
```

```markdown
# 🎡 The Fixed-Trip Theory: for Loop

`for` loop digunakan ketika kita ingin mengeksekusi blok kode untuk **setiap item** dalam sebuah koleksi (Iterable).

### 🏗️ Struktur Dasar
```python
# for [nama_variabel_sementara] in [iterable]:
#     [kode_anak_buah_yang_diulang]

transaksi = [5000, 2500, 10000]

for angka in transaksi:
    print(f"Memproses transaksi sebesar: {angka}")
    # Indentasi di sini berarti baris ini milik si 'for'
```



---

### 🪆 1. Iterasi pada String (Preteli Karakter)
String adalah kumpulan karakter. Kita bisa nge-loop setiap hurufnya.

```python
nama = "FAUZY"
for huruf in nama:
    print(f"Huruf: {huruf}")
```

### 🔢 2. Iterasi dengan Angka (Fungsi `range()`)
Kalau lo mau ngulang sesuatu sebanyak `X` kali tanpa punya list-nya, lo pake `range()`.
* `range(5)` → bakal jalan 5 kali (mulai dari 0 sampai 4).

```python
for i in range(3):
    print(f"Pengecekan ke-{i+1}")
```

---

## 🏗️ Logic Inside Loop (The Decision Matrix Combo)
Di sinilah ilmu **Indentasi** lo diuji. Kita bisa masukin `if` di dalem `for`.

**Skenario:** Lo mau nge-filter transaksi mana yang "Mencurigakan" (di atas 7000).

```python
daftar_transaksi = [5000, 2500, 12000, 3000, 9500]

for nominal in daftar_transaksi:
    # --- LEVEL 1 INDENTASI (Anak buah for) ---
    if nominal > 7000:
        # --- LEVEL 2 INDENTASI (Anak buah if yang ada di dalem for) ---
        print(f"⚠️ ALERT: Transaksi {nominal} terlalu besar!")
    else:
        print(f"✅ Aman: {nominal}")

print("--- Proses Selesai ---") # Luar indentasi = Jalan cuma sekali di akhir
```



---

## 📑 SOP Menulis Loop yang Rapi

1.  **Siapkan Iterable:** Pastikan data lo bisa "dipreteli" (List, String, atau Range).
2.  **Variabel Sementara:** Kasih nama yang masuk akal. Kalau list-nya `transaksi`, variabel sementaranya `nominal`. Kalau list-nya `users`, panggil `user`.
3.  **Jaga Pagar Indentasi:**
    * Satu kali masuk (4 spasi) = Jalan setiap kali loop muter.
    * Dua kali masuk = Jalan cuma kalau `if` di dalem loop bernilai `True`.
    * Sejajar dengan `for` = Jalan cuma sekali setelah loop bener-bener beres.
