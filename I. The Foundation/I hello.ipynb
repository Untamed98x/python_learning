# 📔 How to Use Jupyter (.ipynb) Like a Pro

### ⌨️ Shortcut "Dewa" (Mode Biru/Command Mode)
Tekan **`Esc`** sampai pinggiran sel jadi warna biru, lalu gunakan tombol ini:

* **`A`** : Bikin sel baru di **Atas** (*Above*).
* **`B`** : Bikin sel baru di **Bawah** (*Below*).
* **`M`** : Ubah sel jadi **Markdown** (buat nulis teks).
* **`Y`** : Ubah sel balik jadi **Code** (buat ngetik Python).
* **`D, D`** (tekan D 2x): **Hapus** sel yang dipilih.
* **`Shift + Enter`** : Jalanin sel dan langsung pindah ke sel bawahnya.

---

### 🚠 Cara Geser Urutan Sel (Naik/Turun)

Kalau lo mau mindahin sel Markdown biar ke atas atau sebaliknya:

1.  **Pakai Mouse (VS Code):**
    * Arahkan kursor ke samping kiri sel sampai muncul ikon **titik-titik (Drag Handle)** atau tanda panah **Up/Down**.
    * Tinggal klik dan tarik (*drag*) ke posisi yang lo mau.
2.  **Pakai Shortcut (VS Code):**
    * Tekan **`Option + Panah Atas`** (Mac) atau **`Alt + Panah Atas`** (Windows) buat geser sel ke atas.
    * Tekan **`Option + Panah Bawah`** (Mac) atau **`Alt + Panah Bawah`** (Windows) buat geser sel ke bawah.

---

### 🎨 Tips Format Markdown (Biar Catatan Lo Rapi)

| Hasil | Kode Markdown |
| :--- | :--- |
| **Teks Tebal** | `**Teks Tebal**` |
| *Teks Miring* | `*Teks Miring*` |
| `Code Snippet` | `` `Code Snippet` `` |
| Garis Pembatas | `---` |
| Turun Baris | Beri spasi 2x di akhir baris atau pake `<br>` |
